```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.linear1 = nn.Linear(2, 32)
        self.linear2 = nn.Linear(32, 1)
        
    def forward(self, x):
        x = torch.tanh(self.linear1(x))
        x = self.linear2(x)
        return x  
    
# 
def train(model, optimizer, E, iteration, x, y):
    # 
    losses = []
    for i in range(iteration):
        optimizer.zero_grad()                  # 
        y_pred = model(x)                      # 
        loss = E(y_pred.reshape(y.shape), y)   # 
        loss.backward()                        # 
        optimizer.step()                       # 
        losses.append(loss.item())             # 
    return model, losses

def test(model, x):
    y_pred = model(x).data.numpy().T[0] #
    return y_pred

# 
def plot(x, y, x_new, y_pred, losses):
    # -------------------------------------------------
    # 
    fig = plt.figure(figsize=(9, 4))
    ax1 = fig.add_subplot(121)
    ax2 = fig.add_subplot(122)
 
    # 
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    ax2.set_xlabel('Iteration')
    ax2.set_ylabel('E')
 
    # 
    ax2.set_yscale('log')
 
    # 
    ax1.scatter(x, y, label='dataset')
    ax1.plot(x_new, y_pred, color='red', label='PyTorch regression', marker="o", markersize=3)
    ax2.plot(np.arange(0, len(losses), 1), losses)
    ax2.text(600, 20, 'Training Error=' + str(round(losses[len(losses)-1], 2)), fontsize=16)
    ax2.text(600, 50, 'Iteration=' + str(round(len(losses), 1)), fontsize=16)
 
    # 
    ax1.legend()
    fig.tight_layout()
    plt.grid()
    plt.show()
    plt.close()
    # -------------------------------------------------------------------
    
np.random.seed(2022)
iters = 10000
    
# 
x = np.random.uniform(-10, 10, 100)                                   # x
y = x * x                                                        
x = torch.from_numpy(x.astype(np.float32)).float()                  # x
y = torch.from_numpy(y.astype(np.float32)).float()                  # y
X = torch.stack([torch.ones(100), x], 1)                            # x1

# 
x_test = np.linspace(-15, 15, 60)                                    # x
x_test = torch.from_numpy(x_test.astype(np.float32)).float()        # x
X_test = torch.stack([torch.ones(60), x_test], 1)                   # x1

# 
net = Net()

# 
optimizer = optim.Adam(net.parameters(), lr=0.01)                # Adam
E = nn.MSELoss()   

# 
net, losses = train(model=net, optimizer=optimizer, E=E, iteration=iters, x=X, y=y)

# 
y_pred = test(net, X_test)

# 
plot(x, y, X_test.data.numpy().T[1], y_pred, losses)

extractor = torchvision.models.feature_extraction.create_feature_extractor(net, ["x", "linear1", "tanh", "linear2"])

W1 = net.linear1.weight
W1 = W1.cpu().detach().numpy().copy()
W1 = W1.T
W2 = net.linear2.weight
W2 = W2.cpu().detach().numpy().copy()
W2 = W2.T

U_dashs = []

for V in X:
    features = extractor(V)
    H = features["tanh"]
    H = H.cpu().detach().numpy().copy()
    tanh_dash = (1 - H * H)
    U_dash = np.dot(tanh_dash * W1, W2)
    U_dashs.append(U_dash[1, 0])

plt.scatter(x, U_dashs, color="g")
plt.xlabel("")
plt.ylabel("g'()")
plt.grid()
plt.show()
```

```python
sigma_u = 1
vp = 3
sigma_p = 1
infer_time = 5
epoch = 500
dt = 0.1
phi = list(range(infer_time))
ep = list(range(infer_time))
eu = list(range(infer_time))

# v = [np.random.randint(1, 5) for i in range(epoch)]
v = 4
u = v * v
```

```python
for i in range(epoch):
    phi[0] = vp
    ep[0] = 0
    eu[0] = 0
    
    # 
    for j in range(1, infer_time):
        _phi = np.array([phi[j-1]])
        _phi = np.insert(_phi, 0, 1)
        _phi = torch.from_numpy(_phi.astype(np.float32)).float()
        features = extractor(_phi)
        H = features["tanh"]
        H = H.cpu().detach().numpy().copy()
        tanh_dash = (1 - H * H)
        U = test(net, _phi)
        U_dash = np.dot(tanh_dash * W1, W2)
        U_dash = U_dash[1, 0]
        phi[j] = phi[j-1] + dt * (-ep[j-1] + eu[j-1] * U_dash)
        ep[j] = ep[j-1] + dt * (phi[j-1] - vp - sigma_p * ep[j-1])
        eu[j] = eu[j-1] + dt * (u - U - sigma_u * eu[j-1])
        
    if i == 0:
        plt.scatter(range(infer_time), phi)
        plt.xlabel("infer_time")
        plt.ylabel("")
        plt.plot()
        plt.grid()
        plt.show()

        print("v:", v)
        print("vp:", vp)
        print("sigma_p:", sigma_p)
        print("sigma_u:", sigma_u)
        print("phi:", phi[-1])
        print("u:", u)
        print("U:", U)
          
    # 
    vp = vp + ep[-1] * dt
    sigma_p = max(sigma_p + 1/2 * (ep[-1] ** 2 - 1/sigma_p) * dt, 1)
    sigma_u = max(sigma_u + 1/2 * (eu[-1] ** 2 - 1/sigma_u) * dt, 1) 

    if i == epoch - 1:
        plt.scatter(range(infer_time), phi)
        plt.ylim(-1, 10)
        plt.xlabel("infer_time")
        plt.ylabel("")
        plt.plot()
        plt.grid()
        plt.show()
    
        print("v:", v)
        print("vp:", vp)
        print("sigma_p:", sigma_p)
        print("sigma_u:", sigma_u)
        print("phi:", phi[-1])
        print("u:", u)
        print("U:", U)
```
