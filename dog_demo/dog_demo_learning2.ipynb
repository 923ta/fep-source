```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#　条件設定
dt = 0.01

v_sample = np.random.randint(1, 11, (2, 100))
u_sample = v_sample ** 2

# v_sample = v_sample / np.tile(np.sum(v_sample, axis=0), (2, 1))
# u_sample = u_sample / np.tile(np.sum(u_sample, axis=0), (2, 1))
```

```python
sigma_p = np.cov(v_sample)
sigma_u = np.cov(u_sample)
vp = np.mean(v_sample, axis=1, keepdims=True)
print(sigma_p)
print(sigma_u)
print(vp)
```

```python
theta= np.array([[9, 2],[1, 8]])
```

```python
u = np.array([[10], [10]])
```

```python
v = list(np.array([[0], [0]]) for i in range(500))
v[0] = vp
ep = list(np.array([[0], [0]]) for i in range(500))
eu = list(np.array([[0], [0]]) for i in range(500))
```

```python
for j in range(100):
    # 知覚
    for i in range(1, 500):
        h = v[i-1] * v[i-1]
        g = np.dot(theta, h)
        dg = 2 * np.dot(theta, v[i-1])

        v[i] = v[i-1] + dt * (np.dot(dg.T, eu[i-1]) - ep[i-1])
        ep[i] = ep[i-1] + dt * (v[i-1] - vp - np.dot(sigma_p, ep[i-1]))
        eu[i] = eu[i-1] + dt * (u - g - np.dot(sigma_u, eu[i-1])) 
        
    Hap = list(v[i][0, 0] for i in range(500))
    Ang = list(v[i][1, 0] for i in range(500))
    fig = plt.figure()
    plt.plot(np.arange(0,5,0.01), Hap, color='r', label='Happiness')
    plt.plot(np.arange(0,5,0.01), Ang, color='g', label='Anger')
    plt.xlabel('time')
    plt.legend()
    plt.grid()
    plt.show()

    sp_rev = np.linalg.inv(sigma_p)
    su_rev = np.linalg.inv(sigma_u)

    # 学習
    vp = vp + dt * ep[-1]
    sigma_p = sigma_p + 1/2 * (np.dot(ep[-1], ep[-1].T) - sp_rev) * dt
    sigma_u = sigma_u + 1/2 * (np.dot(eu[-1], eu[-1].T) - su_rev) * dt
    theta = theta + np.dot(eu[-1], h.T) * dt
    
    v[0] = v[-1]
    ep[0] = ep[-1]
    eu[0] = eu[-1]
    
    print(vp)
    print(sigma_p)
    print(sigma_u)
    print(theta)
```
