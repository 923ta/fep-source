```python
#　条件設定
dt = 0.01

# v_sample = np.zeros((2, 100))
# v_sample[: , 0:1] = np.array([[9], [2]])

# for i in range(1, 100):
#     a = np.random.randint(1, 11)
#     if v_sample[0, i-1] == 1:
#         v_sample[0, i] = 2
#         v_sample[1, i] = 9
#     elif v_sample[0, i-1] == 10:
#         v_sample[0, i] = 9
#         v_sample[1, i] = 2
#     else:
#         if a > 5:
#             v_sample[0, i] = v_sample[0, i-1] + 1
#             v_sample[1, i] = v_sample[1, i-1] - 1
#         else:
#             v_sample[0, i] = v_sample[0, i-1] - 1
#             v_sample[1, i] = v_sample[1, i-1] + 1
v_sample = np.random.randint(1, 11, (2, 100))
u_sample = v_sample ** 2
```

```python
sigma_p = np.cov(v_sample)
sigma_u = np.cov(u_sample)
vp = np.mean(v_sample, axis=1, keepdims=True)
print(sigma_p)
print(sigma_u)
print(vp)
```

```python
theta = np.array([[0, 0],[0, 0]])
```

```python
v = list(np.array([[0], [0]]) for i in range(10))
v[0] = vp
ep = list(np.array([[0], [0]]) for i in range(10))
eu = list(np.array([[0], [0]]) for i in range(10))
```

```python
for j in range(100):
    # 知覚
    for i in range(1, 10):
        h = np.dot(v[i-1].T, v[i-1])
        h = h[0, 0]
        g = np.dot(theta, h)
        dg = 2*np.dot(theta, v[i-1])

        v[i] = v[i-1] + dt * (np.dot(dg.T, eu[i-1]) - ep[i-1])
        ep[i] = ep[i-1] + dt * (v[i-1] - vp - np.dot(sigma_p, ep[i-1]))
        eu[i] = eu[i-1] + dt * (u - g - np.dot(sigma_u, eu[i-1])) 
        
    Hap = list(v[k][0, 0] for k in range(10))
    Ang = list(v[k][1, 0] for k in range(10))
    fig = plt.figure()
    plt.plot(np.arange(0,0.1,0.01), Hap, color='r', label='Happiness')
    plt.plot(np.arange(0,0.1,0.01), Ang, color='g', label='Anger')
    plt.xlabel('time')
    plt.legend()
    plt.grid()
    plt.show()

    sp_rev = np.linalg.inv(sigma_p)
    su_rev = np.linalg.inv(sigma_u)

    # 学習
    vp = vp + dt * ep[-1]
    sigma_p = sigma_p + 1/2 * (ep[-1].dot(ep[-1].T) - sp_rev) * dt
    sigma_u = sigma_u + 1/2 * (eu[-1].dot(eu[-1].T) - su_rev) * dt
    theta = theta + eu[-1].dot(h.T) * dt
    
    v[0] = v[-1]
    ep[0] = ep[-1]
    eu[0] = eu[-1]
    
    
    print(u)
    print(vp)
    print(sigma_p)
    print(sigma_u)
    print(theta)
```
