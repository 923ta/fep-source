```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_size, hidden_size, output_size = 1, 10, 1
learning_rate = 0.1
np.random.seed(0)
W1 = np.random.randn(input_size, hidden_size) / np.sqrt(input_size) * np.sqrt(2)
W2 = np.random.randn(hidden_size, output_size) / np.sqrt(hidden_size) * np.sqrt(2)
epoch = 1000
N = 50
```

```python
def relu(x):
    return x * (x > 0)

def softmax(x):
    e = np.exp(x)
    return e / np.sum(e)

def forward(V):
    global W1
    global W2
    H = relu(np.dot(V, W1))
    U = np.dot(H, W2)
    return U, H

def backward(V, H, U, D):
    global W1
    global W2
    dE2 = 2 * (U - D)
    dU = H.reshape(len(H), 1)
    dW2 = dU * dE2
#     dE1 = np.sum(dW2, axis=1)
#     dR = H > 0
#     dH = V.reshape(len(V), 1)
#     dER = dE1*dR
#     dW1 = np.dot(dH, dER.reshape(1, len(dER)))

    dR = H > 0
    dE1 = np.dot(dE2, W2.reshape(1, len(W2))) * dR
    dW1 = np.dot(V.reshape(len(V), 1), dE1.reshape(1, len(dE1)))
    
    W2 -= learning_rate * dW2
    W1 -= learning_rate * dW1
```

```python
# Vs = np.arange(epoch) / epoch 
# np.random.shuffle(Vs)

for i in range(epoch):
    # 順伝搬
    V = np.random.rand(1)
    U, H = forward(V)

    # 誤差逆伝搬
    D = np.sin(2 * np.pi * V)
    backward(V, H, U, D)
    print(2 * (U - D))
    print("V:", V)
    print("H:", H)
    print("U:", U)
    print("D:", D)
    
Vs = np.arange(100) / 100  
Us = []

for V in Vs:
    # 順伝搬
    U, H = forward(V)
    Us.append(U)
    
# plt.scatter(Vs, Us)    
# plt.show()
```
